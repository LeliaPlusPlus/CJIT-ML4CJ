# Spatial Visualization

In this tutorial, we provide basic code to get started with spatial visualizations, using a urban heat island dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.environ['USE_PYGEOS'] = '0'
import geopandas as gpd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
gdf = gpd.read_file('data/uhi_shp/uhi.shp')
```

```python
gdf.head()
```

```python
fig, ax = plt.subplots(3, 1, figsize = (24,18), constrained_layout=True)
columns = ['Black', 'Hispanic', 'White']
cmap='Purples'
for i in range(3):
    gdf.plot(
            column=columns[i],
            ax=ax[i], 
            legend=True,
            legend_kwds={"orientation": "vertical", "pad": -0.01, "shrink": 0.5},
            cmap = cmap    
    );
for i in range(3):
    ax[i].set_axis_off();
ax[0].set_title('Percent Black', fontsize=16)
ax[1].set_title('Percent Hispanic', fontsize=16)
ax[2].set_title('Percent White', fontsize=16)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (24,24), constrained_layout=True)
cmap='YlOrRd'
gdf.plot(
        column='UHI',
        ax=ax, 
        legend=True,
        legend_kwds={"orientation": "vertical", "pad": -0.01, "shrink": 0.5},
        cmap = cmap    
);
ax.set_axis_off();
ax.set_title('SUHI Distribution by Census Tract',fontsize=20)
```

```python
fig, ax = plt.subplots(figsize = (24,18), constrained_layout=True)

gdf.plot(
        column='FPL200S',
        ax=ax, 
        categorical=True,
        legend=True,
        #legend_kwds={"orientation": "vertical", "pad": -0.01, "shrink": 0.5},
        cmap = 'Spectral'    
);

ax.set_axis_off();
ax.set_title('CEJST Low Income Indicator by Tract', fontsize=30)
```
